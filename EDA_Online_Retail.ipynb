# Exploratory Data Analysis (EDA) — Online Retail Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('online_retail.csv')
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
df.isnull().sum()
```

```python
sns.histplot(df['TotalAmount'], kde=True)
plt.show()
```

```python
df.groupby(df['InvoiceDate'].dt.to_period('M'))['TotalAmount'].sum().plot()
plt.show()
```

```python
df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False).plot(kind='bar')
plt.show()
```

```python
sns.heatmap(df[['Quantity','UnitPrice','TotalAmount']].corr(), annot=True)
plt.show()
```

## Conclusion
- Dataset cleaned and explored successfully.
- Sales distribution is right‑skewed.
- Certain countries contribute more revenue.
- Quantity strongly impacts total sales.
